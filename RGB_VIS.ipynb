```python
'''
Libraries used:
    OpenCV (cv2) - reading and manipulating video file data
    numpy - for random selection
    os - file manipulation
    Matplotlib - plotting data on graphs   
'''
import os
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

seed_constant=23
np.random.seed(seed_constant)
```

```python
# create Matplotlib figure
plt.figure(figsize=(50,60))
```

```python
# create list of folders in RGB folder
all_classes_names=os.listdir('RGB')
```

```python
# keep counter for subplots
counter=1
# loop through class folders
for selected_class in all_classes_names:
    # create files list of videos in class folder
    files_list=os.listdir(f'RGB/{selected_class}')
    #select random file
    selected_file=np.random.choice(files_list)
    #read video file
    video=cv.VideoCapture(f'RGB/{selected_class}/{selected_file}')
    #find the middle frame in video sequence
    mid_frame=video.get(cv.CAP_PROP_FRAME_COUNT)//2
    #set video to middle frame because first frame images do not distinguish actions 
    video.set(cv.CAP_PROP_POS_FRAMES, mid_frame-1)
    _,rgb_frame=video.read()
    #close object and release all resources
    video.release()
    #put text on video
    cv.putText(rgb_frame,selected_class,(60,100),cv.FONT_HERSHEY_SIMPLEX,2,(255,0,0),2)
    # plot subplot on main figure
    plt.subplot(5,6,counter)
    plt.imshow(rgb_frame)
    plt.axis('off')
    counter+=1
```
